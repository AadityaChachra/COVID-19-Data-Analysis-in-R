#### Objective:
This project aims to analyze and visualize the COVID-19 dataset from **Our World in Data** to explore trends, distributions, and comparisons between China and India. The dataset includes a variety of metrics such as total cases, new cases, deaths, vaccination rates, and socioeconomic factors.

#### Description:
Using the powerful capabilities of the **R programming language**, this project focuses on extracting meaningful insights from a large and complex COVID-19 dataset. Key aspects include:

1. **Data Processing**:  
   - Loading the dataset from the provided link.  
   - Filtering data for specific countries (China and India) to create a comparative analysis.  
   - Handling missing values to ensure accurate visualizations.  

2. **Visualizations**:  
   A comprehensive set of plots was generated to analyze the following metrics:  
   - **COVID-19 Trends**: Total cases, new cases, deaths, and reproduction rates over time.  
   - **Healthcare Metrics**: Hospital and ICU patients, vaccination rates, and testing rates.  
   - **Socioeconomic Factors**: Life expectancy, stringency index, population density, and diabetes prevalence.  
   - **Lifestyle Indicators**: Smoking prevalence among males and females.  

3. **Graphical Enhancements**:  
   Each visualization is tailored for clarity, featuring:  
   - Intuitive labeling of axes and titles.  
   - Color-coding to distinguish between China and India.  
   - Improved readability with larger fonts and centered titles.  

4. **Comparative Insights**:  
   Visualizations provide insights into the differences and similarities in the pandemic's impact, healthcare response, and demographic factors between the two countries.

#### Tools and Libraries:
- **Libraries Used**: `ggplot2`, `dplyr`, `tidyr`, and `readr`.  
- **Data Source**: Open-source COVID-19 dataset from (https://covid.ourworldindata.org/data/owid-covid-data.csv).  

#### Outcomes:
This project offers a data-driven exploration of the pandemic's trajectory in China and India, highlighting key patterns and disparities. The visualizations help policymakers, researchers, and the general public understand the pandemic's impact and the effectiveness of mitigation efforts.

```r
# Load necessary libraries
library(ggplot2)
library(dplyr)
library(readr)
library(tidyr)

# Read the dataset
data <- read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")

# Convert the 'date' column to Date format
data$date <- as.Date(data$date)

china_data <- data %>% filter(location == "China")

# Plot the total cases over time
ggplot(china_data, aes(x = date, y = total_cases)) +
  geom_line() +
  labs(title = "COVID-19 Total Cases Over Time in the China",
       x = "Date",
       y = "Total Cases") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# Filter the data for India
india_data <- data %>% filter(location == "India")

# Plot the total cases over time for India
ggplot(india_data, aes(x = date, y = total_cases)) +
  geom_line(color = "red") +  # Use a different color for India's plot
  labs(title = "COVID-19 Total Cases Over Time in India",
       x = "Date",
       y = "Total Cases") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# Filter the data for the China and India
combined_data <- data %>% filter(location %in% c("China", "India"))

# Plot the data
ggplot(combined_data, aes(x = date, y = total_cases, color = location)) +
  geom_line(size = 1.2) +  # Adjust line size for better visibility
  labs(title = "COVID-19 Total Cases Over Time: China vs. India",
       x = "Date",
       y = "Total Cases",
       color = "Country") +  # Legend title
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# Additional Combined Graphs
# 1. New Cases Over Time
ggplot(combined_data, aes(x = date, y = new_cases, color = location)) +
  geom_line(size = 1) +
  labs(title = "New COVID-19 Cases Over Time: China vs. India",
       x = "Date",
       y = "New Cases",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 2. Total Deaths Over Time
ggplot(combined_data, aes(x = date, y = total_deaths, color = location)) +
  geom_line(size = 1) +
  labs(title = "Total COVID-19 Deaths Over Time: China vs. India",
       x = "Date",
       y = "Total Deaths",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 3. New Deaths Per Million
ggplot(combined_data, aes(x = date, y = new_deaths_per_million, color = location)) +
  geom_line(size = 1) +
  labs(title = "New Deaths Per Million: China vs. India",
       x = "Date",
       y = "New Deaths Per Million",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 4. Total Vaccinations Per Hundred
ggplot(combined_data, aes(x = date, y = total_vaccinations_per_hundred, color = location)) +
  geom_line(size = 1) +
  labs(title = "Total Vaccinations Per Hundred: China vs. India",
       x = "Date",
       y = "Total Vaccinations Per Hundred",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 5. Reproduction Rate Over Time
ggplot(combined_data, aes(x = date, y = reproduction_rate, color = location)) +
  geom_line(size = 1) +
  labs(title = "COVID-19 Reproduction Rate Over Time: China vs. India",
       x = "Date",
       y = "Reproduction Rate",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 6. Stringency Index Over Time
ggplot(combined_data, aes(x = date, y = stringency_index, color = location)) +
  geom_line(size = 1) +
  labs(title = "Stringency Index Over Time: China vs. India",
       x = "Date",
       y = "Stringency Index",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 7. New Cases Smoothed
ggplot(combined_data, aes(x = date, y = new_cases_smoothed, color = location)) +
  geom_line(size = 1) +
  labs(title = "Smoothed New COVID-19 Cases: China vs. India",
       x = "Date",
       y = "New Cases Smoothed",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 8. Total Cases Per Million
ggplot(combined_data, aes(x = date, y = total_cases_per_million, color = location)) +
  geom_line(size = 1) +
  labs(title = "Total Cases Per Million: China vs. India",
       x = "Date",
       y = "Total Cases Per Million",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 9. New Cases Per Million
ggplot(combined_data, aes(x = date, y = new_cases_per_million, color = location)) +
  geom_line(size = 1) +
  labs(title = "New Cases Per Million: China vs. India",
       x = "Date",
       y = "New Cases Per Million",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 10. Total Deaths Per Million
ggplot(combined_data, aes(x = date, y = total_deaths_per_million, color = location)) +
  geom_line(size = 1) +
  labs(title = "Total Deaths Per Million: China vs. India",
       x = "Date",
       y = "Total Deaths Per Million",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 11. Hospital Patients Over Time
ggplot(combined_data, aes(x = date, y = hosp_patients, color = location)) +
  geom_line(size = 1) +
  labs(title = "Hospital Patients Over Time: China vs. India",
       x = "Date",
       y = "Number of Hospital Patients",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 12. ICU Patients Per Million
ggplot(combined_data, aes(x = date, y = icu_patients_per_million, color = location)) +
  geom_line(size = 1) +
  labs(title = "ICU Patients Per Million: China vs. India",
       x = "Date",
       y = "ICU Patients Per Million",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 13. Total Tests Per Thousand
ggplot(combined_data, aes(x = date, y = total_tests_per_thousand, color = location)) +
  geom_line(size = 1) +
  labs(title = "Total Tests Per Thousand: China vs. India",
       x = "Date",
       y = "Total Tests Per Thousand",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 14. Positive Rate Over Time
ggplot(combined_data, aes(x = date, y = positive_rate, color = location)) +
  geom_line(size = 1) +
  labs(title = "Positive Rate Over Time: China vs. India",
       x = "Date",
       y = "Positive Rate",
       color = "Country") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 15. Life Expectancy Comparison
life_expectancy_data <- combined_data %>%
  drop_na(life_expectancy) %>%
  group_by(location) %>%
  summarize(life_expectancy = first(life_expectancy))

if (nrow(life_expectancy_data) > 0) {
  ggplot(life_expectancy_data, aes(x = location, y = life_expectancy, fill = location)) +
    geom_bar(stat = "identity", width = 0.5) +
    labs(title = "Life Expectancy Comparison: China vs. India",
         x = "Country",
         y = "Life Expectancy") +
    theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)} else {
  print("No data available for life expectancy.")
}

# 16. Human Development Index (HDI) Comparison
hdi_data <- combined_data %>%
  filter(!is.na(human_development_index)) %>%  # Remove rows with NA values
  group_by(location) %>%
  summarize(human_development_index = first(human_development_index))  # Take the first available value

if (nrow(hdi_data) > 0) {
  ggplot(hdi_data, aes(x = location, y = human_development_index, fill = location)) +
    geom_bar(stat = "identity", width = 0.5) +
    labs(title = "Human Development Index (HDI): China vs. India",
         x = "Country",
         y = "Human Development Index") +
    theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)} else {
  print("No data available for Human Development Index.")
}

# 17. Population Density
population_density_data <- combined_data %>%
  filter(!is.na(population_density)) %>%  # Remove rows with NA values
  group_by(location) %>%
  summarize(population_density = first(population_density))

ggplot(population_density_data, aes(x = location, y = population_density, fill = location)) +
  geom_bar(stat = "identity", width = 0.5) +
  labs(title = "Population Density: China vs. India",
       x = "Country",
       y = "Population Density") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 18. Diabetes Prevalence
diabetes_data <- combined_data %>%
  filter(!is.na(diabetes_prevalence)) %>%  # Remove rows with NA values
  group_by(location) %>%
  summarize(diabetes_prevalence = first(diabetes_prevalence))

ggplot(diabetes_data, aes(x = location, y = diabetes_prevalence, fill = location)) +
  geom_bar(stat = "identity", width = 0.5) +
  labs(title = "Diabetes Prevalence: China vs. India",
       x = "Country",
       y = "Diabetes Prevalence (%)") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 19. Female Smokers
female_smokers_data <- combined_data %>%
  filter(!is.na(female_smokers)) %>%  # Remove rows with NA values
  group_by(location) %>%
  summarize(female_smokers = first(female_smokers))

ggplot(female_smokers_data, aes(x = location, y = female_smokers, fill = location)) +
  geom_bar(stat = "identity", width = 0.5) +
  labs(title = "Female Smokers (%): China vs. India",
       x = "Country",
       y = "Female Smokers (%)") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)

# 20. Male Smokers
male_smokers_data <- combined_data %>%
  filter(!is.na(male_smokers)) %>%  # Remove rows with NA values
  group_by(location) %>%
  summarize(male_smokers = first(male_smokers))

ggplot(male_smokers_data, aes(x = location, y = male_smokers, fill = location)) +
  geom_bar(stat = "identity", width = 0.5) +
  labs(title = "Male Smokers (%): China vs. India",
       x = "Country",
       y = "Male Smokers (%)") +
  theme_minimal(base_size = 14) +  # Increase font size for better readability
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold"),  # Center the title
    legend.position = "top"  # Position the legend at the top
)
```
